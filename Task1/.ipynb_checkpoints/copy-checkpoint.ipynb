#### 3. Классификация методом ближайшего соседа

Ранее в контесте вы уже реализовали функцию поиска ближайших соседей. Теперь давайте применим ее для решения нашей текущей задачи. 

Реализуйте функцию, которая для каждого объекта тестового набора данных найдет $k$ ближайших соседей из обучающего набора данных, и в качестве ответа возьмет наиболее часто встречающееся значение целевого признака среди найденных соседей. Обратите внимание, что соседей необходимо искать только среди объектов обучающего набора данных.

```python
def knn(features, train_features, train_target, k=1):
    '''
    Предсказание целевого признака для данных features -- pandas-таблица данных.
    на основе обучающего набора данных train_features, train_target.
    Возвращает pandas.Series с теми же индексами, что и у features.
    Число k -- количество соседей.
    '''
    # Поиск расстояния
    train_features = train_features[np.newaxis, :]
    features = features[:, np.newaxis]
    dist = np.sqrt(np.sum((train_features - features) ** 2, axis=2))
    
    # Нахождение наиболее часто встречающегося признака
    neighbour = dist.argsort(axis=1)[:, :k]
    result = np.array([0]*features.shape[0])
    for i in range(len(result)):
        result[i] = np.take(train_target, neighbour[i]).sum()
    ...
    return 1 * (result > k / 2)
```

Постройте классификатор методом ближайшего соседа, используя только признаки `Age` и `Fare`. 

Какую метрику (функцию расстояния) вы при этом используете? Нужно ли предварительно что-то сделать с признаками?

Предварительно нужно обработать цены за билет, чтобы они вносили вклад в метрику в соответствии с вкладов возраста. Причем в силу проведенного наблюдения, можно изменить вклад возраста: чем младше, тем больший вклад

```python
# Функция, принимает dataframe и возвращает обдработанный для метрики dataframe
def change_for_metrics(df):
    changed_df = df[["Age", "Fare"]]
    changed_df["Age"] = changed_df["Age"] / changed_df["Age"].mean()
#     changed_df["Age"] = 1 / changed_df["Age"]
    
    changed_df["Fare"] = changed_df["Fare"] / changed_df["Fare"].mean()
    return changed_df
```

Как меняется результат классификации при другом количестве соседей? Рассмотрите хотя бы три различных нечетных значения количества соседей.

```python
# Подготовка данных
test_changed = change_for_metrics(features_test).to_numpy()
train_changed = change_for_metrics(features_train).to_numpy()
target = target_train.to_numpy()
```

```python
# Cоздадим np.array результатов knn для разных k
test_sz = len(target_test)
knn_results = np.array([0] * test_sz)
for i in range(1, test_sz + 1):
    result = knn(test_changed, train_changed, target, i)
    knn_results[i - 1] = np.sum(result == target_test)
knn_results = knn_results / test_sz
```

```python
# Визуализация данных
knn_df = pd.DataFrame({"Results": knn_results, 
                       "Indexes": np.array(range(test_sz))})

plt.figure(figsize=(15, 5))
sns.lineplot(data=knn_df, x="Indexes", y="Results")
knn_results.mean()
```

Результаты работы алгоритма knn показывают, что нужно брать параметр k больше 5, в промежутке 5-40 наблюдается хаотичное изменения в районе 0.7, максимум на промежутке 50-80, а дальше небольшое убывание и стабтлтзация почти до конца

```python
# Рандомно выбираем 5 нечетных числа 
indexes = np.random.randint(1, test_sz / 2, 5)
indexes = indexes * 2 - 1

print(indexes)
print(knn_results[indexes - 1])
```

**Дерево 1:**

Процент выживаемости среди женщин около 75%, также среди мужчин наивысщая доля выживших наблюдается среди детей. Это дерево, опираясь на эти знания, предсказывает, что женщины и дети мужского пола выживут.

```python
def tree1(features):
    '''
    Предсказание целевого признака для данных features -- pandas-таблица данных.
    Возвращает pandas.Series с теми же индексами, что и у features.
    '''
    predicted = features["Sex"] == "female"
    predicted = predicted | (features["Age"] < 15)
    return predicted
```

```python
def tree2(features):
    '''
    Предсказание целевого признака для данных features -- pandas-таблица данных.
    Возвращает pandas.Series с теми же индексами, что и у features.
    '''
    return features["Pclass"] != 3
```

#### 5. Качество

Вспомним, что у нас имеется тестовая часть выборки. Самое время ее использовать для того, чтобы оценить, насколько хорошими получились деревья. Предложите какой-нибудь критерий качества.

*Замечание*. Не стоит пытаться искать, какие критерии существуют. В данном случае легко предложить адекватный критерий. Стандартные критерии мы разберем позже, но для начала лучше подумать самостоятельно.

**Критерий:**

Критерием будет процент правильно вычисленных предсказаний

```python
def criteria(target, predicted):
    '''
    Вычисление критерия качества для предсказания predicted, 
    если истинные значения -- target.
    Возвращает одно вещественное число.
    '''
    return np.sum(target == predicted) / target.shape[0]
```

```python
criteria(target_test, tree1(features_test))
```
